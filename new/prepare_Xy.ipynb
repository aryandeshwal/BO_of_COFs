# import and prepare data.

```python
import numpy as np
import pickle
import pandas as pd
```

load data from Mercado et al., drop outlier

```python
df = pd.read_csv('properties.csv')
# we remove row 48225 since it is an outlier for void fraction feature  # can be seen by df[df[' void fraction [widom]'] > 1]
df = df.drop(48225)
df.head()
```

define new feature as density of elements

```python
elements = ['carbon', 'fluorine', 'hydrogen', 'nitrogen', 'oxygen', 'sulfur', 'silicon']
for el in elements:
    df["density of " + el] = df[' num ' + el] / df[' supercell volume [A^3]']
df.head()
```

get feature matrix and target vector

```python
features = [' void fraction [widom]', ' density [kg/m^3]', ' largest included sphere diameter [A]', ' largest free sphere diameter [A]', ' surface area [m^2/g]',
            'density of carbon', 'density of fluorine', 'density of hydrogen', 'density of nitrogen', 'density of oxygen', 'density of sulfur', 'density of silicon']
print("# features: ", len(features))
```

```python
y = df[' deliverable capacity [v STP/v]'].values
print("shape of Y: ", np.shape(y))
y
```

```python
X = df[features].values
print("shape of X: ", np.shape(X))
X
```

Min-Max normalize the features so that they lie in $[0, 1]$. this is ok to do over all data as opposed to just training because in BO we will compute the cheap features of every COF in the database.

```python
for i in range(np.shape(X)[1]):
    X[:, i] = (X[:, i] - np.min(X[:, i])) / (np.max(X[:, i]) - np.min(X[:, i]))
    print("feature", i, " in [", np.min(X[:, i]), ",", np.max(X[:, i]), "]")
X
```

```python
with open('inputs_and_outputs.pkl', 'wb') as file:
    pickle.dump({'X': X, 'y': y, 'features': features}, file)
```
